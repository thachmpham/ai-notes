# Plot

## Load data

````python
import pandas as pd
import matplotlib.pyplot as plt

college = pd.read_csv("../data/College.csv")
print(*college.columns, sep=',\t')
````

## Scatter plot

````python
college.plot.scatter('Apps', 'Accept')
````

## Histogram

### How many private schools ?

````python
private = college['Private'].astype('category')
private
````

````python
private.value_counts()
````

````python
private.hist()
````

### How many schools has PhD number in range 0-10+ ?

````python
college['PhD'].value_counts(bins=10).sort_index()
````

**Explain output**  
```
(7.904, 17.5]      5
```
How many schools has PhD number in range (7.904, 17.5] ?  
Answer: 5 schools

````python
college.hist('PhD', bins=10)
````
