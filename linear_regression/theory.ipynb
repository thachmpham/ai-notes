# Simple Linear Regression
**Predict a quantitative response Y on a single predictor variable X.**

## Definitions
### True relationship (y)
```
y ≈ β0 + β1 x
```

### Predict (ŷ)
```
ŷi = β̂0 + β̂1
```

### Coefficients
```
ŷi = β̂0 + β̂1 x

β̂0: intercept  
β̂1: slope
```

### Mean, Average or Expected value (µ : 'mu', E[X]) 
The middle value of the data set.

```
E[X] = sum(X) / n
            n:      number of element

or

E[X] = sum(xi * Pr(xi))
            xi:     value of the i-th element
            Pr(xi): probability of xi
```

### Residual (e)
The diﬀerence between prediction and sample  
```
ei = yi − ŷi
```

### Residual Sum of Square (RSS)
```
RSS =   e1^2 + e2^2 + .. en^2
```
            

### Standard Deviation (σ: 'sigma')
How much the data is spread out around the average.  

```
σ = sqrt(RSS)
```

68%     data in     µ +/- σ
95%     data in     µ +/- 2σ
99.7%   data in     µ +/- 3σ

### Standard Error (SE)

```
SE = σ / √n
```

As the sample size becomes larger, the standard error will become smaller, indicating that the estimated mean value become better approximates the population mean.  

The smaller the standard error the better. A standard error of zero (or close to it) would indicate that the estimated value is exactly the true value.  

Used to compute confidence intervals and to statistically test the hypothesis of the existence of a relationship.  

### 95% Confidence Interval (CI)
CI = 95% meaning:
- 95% values around mean are true.
- 05% values in two corners are false.


### Correlation coefficient (p)
A correlation coefficient measures the relationship between two variables.
- Range: -1 <= p <= 1
- p close to 1,-1: there's relation
- p close to 0: no significant relation

### Correlation & Confidence Interval
```
if Correlation <= Unconfidence Interval:
    there no relation
```
Confidence interval 95%:  
=> Confidence interval = 0.95  
=> Unconfidence interval = 0.05  

if Correlation = 0.04 < 0.05 = Unconfidence:  
=> there no relation  
    
if Correlation = 0.05 == 0.05 = Unconfidence:  
=> there no relation  
   
if Correlation = 0.06 > 0.05 = Unconfidence:  
=> there relation

### Correlation test (p-value)
Unlike a correlation matrix which indicates the correlation coefficients between some pairs of variables in the **sample**, a correlation test is used to test whether the correlation (denoted ρ) between 2 variables is significantly different from 0 or not in the **population**.
```
H0: ρ = 0 (meaning that there is no linear relationship between the two variables)
H1: ρ ≠ 0 (meaning that there is a linear relationship between the two variables)
```
